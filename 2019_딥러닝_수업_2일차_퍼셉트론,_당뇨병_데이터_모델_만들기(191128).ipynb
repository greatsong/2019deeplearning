```
# 퍼셉트론
w1 = 1
w2 = 1
b = 1

def perceptron(x1, x2) : 
  y = w1 * x1 + w2 * x2 + b
  if y > 0:
    return 1
  else:
    return 0

for x1, x2 in [(0, 0), (0, 1), (1, 0), (1, 1)]:
  print('입력:',x1, x2, ' 출력:',perceptron(x1, x2))
```

```
# 1단계 : 데이터 준비
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
# 1단계 : 데이터 준비 - 데이터 모양 확인하기
diabetes.data.shape
```

```
# 1단계 : 데이터 준비 - 학습 데이터, 정답 레이블 나누기
x = diabetes.data[:, 2]
y = diabetes.target
```

```
# 1단계 : 데이터 준비 - 데이터 모양 확인하기
x.shape
```

```
# 1단계 : 데이터 준비 - 데이터 모양 확인하기(상관관계)
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```

```
# 2단계 : w, b 초기화

w = 1
b = 1
```

```
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.plot([-0.1, 0.2], [-0.1 * w + b, 0.2 * w + b], 'g')
plt.show()
```

```
# 3단계 : 모델(wx +b)에서 대한 예측값

y2 = x[0] * w + b
```

```
x[0]
```

```
y2
```

```
# 4단계 : 정답과 예측에 대한 w값 업데이트

w2 = w + (y[0] - y2) * x[0]
w2
```

```
# 4단계 : 정답과 예측에 대한 b값 업데이트

b2 = b + (y[0] - y2)
b2
```

```
# 5단계 : 반복하기

for i in range(len(x)) :
  y2 = x[i] * w + b
  err = y[i] - y2
  w = w + err * x[i]
  b = b + err
print(w,b)
```

```
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.plot([-0.1, 0.2], [-0.1 * w + b, 0.2 * w + b], 'g')
plt.show()
```

```
# 5단계 : 더 반복하기(epoch)

for j in range(10) : 
  for i in range(len(x)) :
    y2 = x[i] * w + b
    err = y[i] - y2
    w = w + err * x[i]
    b = b + err
print(w,b)
```

```
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.plot([-0.1, 0.2], [-0.1 * w + b, 0.2 * w + b], 'g')
plt.show()
```

```
# 6단계 : 모델로 예측하기

import numpy as np
x2 = np.array([-0.05, 0.12, 0.15])
y2 = x2 * w + b

plt.scatter(x, y)
plt.scatter(x2, y2, color = 'r')
plt.show()
```
