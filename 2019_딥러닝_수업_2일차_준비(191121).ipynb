```
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
diabetes.data.shape
```

```
x = diabetes.data[:, 2]
y = diabetes.target
```

```
x.shape
```

```
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```

```
w = 1
b = 1
```

```
y2 = x[0] * w + b
```

```
x[0]
```

```
y2
```

```
w2 = w + x[0] * (y[0] - y2)
w2
```

```
b2 = b + 1 * (y[0] - y2)
b2
```

```
for i in range(len(x)) :
  y2 = x[i] * w + b
  err = y[i] - y2
  w = w + x[i] * err
  b = b + 1 * err
print(w,b)
```

```
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.plot([-0.1, 0.2], [-0.1 * w + b, 0.2 * w + b], 'g')
plt.show()
```

```
for _ in range(100) : 
  for i in range(len(x)) :
    y2 = x[i] * w + b
    err = y[i] - y2
    w = w + x[i] * err
    b = b + 1 * err
print(w,b)
```

```
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.plot([-0.1, 0.2], [-0.1 * w + b, 0.2 * w + b], 'g')
plt.show()
```
